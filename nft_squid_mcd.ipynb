## NFT SQUID - M(H)

```python
from matplotlib import pylab as plt
import pandas as pd
import os
from os import listdir
from os.path import isfile, join, realpath
import re
import pandas as pd

mypath = 'D:/sandbox/nft-squid' 
os.chdir(mypath)

files = [f for f in listdir(mypath) if isfile(join(mypath, f)) and f.split(".")[-1] == "dat"]
print(files)

# Load the data into a pandas DataFrame
df = pd.read_csv('crystall_2k.dat', header=27, sep=',')

# View the first 5 rows (default)
#print(df.head(5))  # You can also specify a number inside the parentheses like df.head(10)

for file in files:
    df = pd.read_csv(file, header=27, sep=',')
    #df.plot(x='COM4.Field', y='GPIB0::4::INSTR.x', kind='line')
    x = df['Magnetic Field (Oe)']
    y = df['Moment (emu)']
    plt.plot(x, y, linewidth=1)
    plt.title(file)
    plt.legend()
    plt.ylabel('Moment (emu)') 
    plt.xlabel('H (Oe)')
    plt.show()
    
    selcolumns = df[['Magnetic Field (Oe)','Moment (emu)']]
    output_file = file + '_new' + '.csv'
    #save_path = mypath #r"D:\my_folder\output_file.csv"
    selcolumns.to_csv(output_file, index=False)
```
